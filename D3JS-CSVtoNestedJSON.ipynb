```python
df2013 = pd.DataFrame(df,columns=['Regions','Region_Type','Colour_Code','2013'])
df2014 = pd.DataFrame(df,columns=['Regions','Region_Type','Colour_Code','2014'])
df2015 = pd.DataFrame(df,columns=['Regions','Region_Type','Colour_Code','2015'])
df2016 = pd.DataFrame(df,columns=['Regions','Region_Type','Colour_Code','2016'])
df2017 = pd.DataFrame(df,columns=['Regions','Region_Type','Colour_Code','2017'])

df2013=df2013.rename(index=str,columns={"2013":"value"})
df2014=df2014.rename(index=str,columns={"2014":"value"})
df2015=df2015.rename(index=str,columns={"2015":"value"})
df2016=df2016.rename(index=str,columns={"2016":"value"})
df2017=df2017.rename(index=str,columns={"2017":"value"})

df2013['year'] = '2013'
df2014['year'] = '2014'
df2015['year'] = '2015'
df2016['year'] = '2016'
df2017['year'] = '2017'
```

```python
df_appended = df2013.append(df2014).append(df2015).append(df2016).append(df2017).reset_index(drop=True)
df_appended = df_appended[['year','Region_Type','Colour_Code','Regions','value']]
df_appended.to_csv('df_appended.csv', encoding='utf-8',index=False)
```

```python
nest = dict()
nest = {'name':'flare','data':[]}

for line in df_appended.values:
    year = line[0]
    regiontype = line[1]
    colour = line[2]
    region = line[3]
    value = line[4]
    
    years_list = []
    for item in nest['data']:
        years_list.append(item['year'])
    
    if not year in years_list:
        nest['data'].append({
            'year':year,
            'data':[{
                'name':regiontype,
                'colour':colour,
                'data':[{
                    'name':region,'value':value
                }]
            }]
        })
        
    else:
        nest['data'][years_list.index(year)]['data'].append({
            'year':year,
            'data':[{
                'name':regiontype,
                'colour':colour,
                'data':[{
                    'name':region,'value':value
                }]
            }]
        })
        
```

```python
def yearflare(df, year,filename):
    flare = dict()
    flare = {'year':year,'children':[]}
    
    for line in df.values:
        regiontype = line[0]
        colour = line[1]
        region = line[2]
        value = line[3]
        
        keys_list=[]
        for item in flare['children']:
            keys_list.append(item['name'])
        if not regiontype in keys_list:
            flare['children'].append({
                'name':regiontype,
                'children':[{'name':region,'value':value}]
            })
        else:
            flare['children'][keys_list.index(regiontype)]['children'].append({
                'name':region,
                'value':value
            })
    #return flare
    with open(filename,'w') as outfile:
        json.dump(flare,outfile)
```

```python
yearflare(df2013_m,2013,'flarejson2013.json')
yearflare(df2014_m,2014,'flarejson2014.json')
yearflare(df2015_m,2015,'flarejson2015.json')
yearflare(df2016_m,2016,'flarejson2016.json')
yearflare(df2017_m,2017,'flarejson2017.json')
```
